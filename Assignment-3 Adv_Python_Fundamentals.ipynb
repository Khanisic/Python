```python
data = pd.read_csv('countries.csv')
data
```

```python
bd=data[data.country == 'Bahrain']
bd
```

```python
ger=data[data.country == 'Germany']
ger
```

```python
from matplotlib import pyplot as plt  
```

```python
plt.plot(bd.year,bd.population/10**6)
plt.plot(ger.year,ger.population/10**6)
plt.xlabel('year')
plt.ylabel('polupation')
plt.title('Comparision of Countries')
plt.legend(['This is Bahrain','This is Germany'])
plt.show()
```
